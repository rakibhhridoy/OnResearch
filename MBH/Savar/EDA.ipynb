```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
from scipy.stats import linregress
```

```python
df = pd.read_csv("Data.csv", index_col=False)
```

```python
df
```

```python
sns.pairplot(df)
```

```python
g = sns.pairplot(df, kind='reg', palette="viridis")

# Function to annotate the plots with regression coefficients
def annotate_regression_coefficients(x, y, **kwargs):
    slope, intercept, r_value, p_value, std_err = linregress(x, y)
    ax = plt.gca()
    label = f'r = {r_value:.2f}\nslope = {slope:.2f}'
    ax.annotate(label, xy=(0.1, 0.9), xycoords='axes fraction', ha='left', va='center', 
                bbox=dict(boxstyle='round,pad=0.5', fc='white', alpha=0.5))

# Map the annotation function to the lower triangle
g.map_lower(annotate_regression_coefficients)

plt.show()
```

```python
g = sns.pairplot(df, kind='reg')

# Function to annotate the plots with regression coefficients and p-values
def annotate_regression_coefficients(x, y, **kwargs):
    slope, intercept, r_value, p_value, std_err = linregress(x, y)
    ax = plt.gca()
    label = f'r = {r_value:.2f}\np = {p_value:.2e}'
    ax.annotate(label, xy=(0.1, 0.9), xycoords='axes fraction', ha='left', va='center', 
                bbox=dict(boxstyle='round,pad=0.5', fc='white', alpha=0.5))

# Map the annotation function to the lower triangle
g.map_lower(annotate_regression_coefficients)

plt.show()
```

```python
g = sns.pairplot(df, kind='reg', plot_kws={'scatter_kws': {'color': 'purple'}, 'line_kws': {'color': 'black'}})

# Function to annotate the plots with regression coefficients and p-values
def annotate_regression_coefficients(x, y, **kwargs):
    slope, intercept, r_value, p_value, std_err = linregress(x, y)
    ax = plt.gca()
    label = f'r = {r_value:.2f}\nslope = {slope:.2f}\np = {p_value:.2e}'
    ax.annotate(label, xy=(0.1, 0.9), xycoords='axes fraction', ha='left', va='center', 
                bbox=dict(boxstyle='round,pad=0.5', fc='white', alpha=0.5))

# Map the annotation function to the lower triangle
g.map_lower(annotate_regression_coefficients)

plt.show()
```

```python
g = sns.pairplot(df, kind='reg', plot_kws={'scatter_kws': {'color': 'purple'}, 'line_kws': {'color': 'black'}})

# Function to annotate the plots with regression coefficients, p-values, and the regression equation
def annotate_regression_coefficients(x, y, **kwargs):
    slope, intercept, r_value, p_value, std_err = linregress(x, y)
    ax = plt.gca()
    equation = f'y = {intercept:.2f} + {slope:.2f}x'
    label = f'{equation}\nr = {r_value:.2f}\np = {p_value:.2e}'
    ax.annotate(label, xy=(0.05, 0.9), xycoords='axes fraction', ha='left', va='center', 
                bbox=dict(boxstyle='round,pad=0.5', fc='white', alpha=0.5))

# Map the annotation function to the lower triangle
g.map_lower(annotate_regression_coefficients)

plt.show()
```

```python
height = 3.5
aspect = 1.2
g = sns.pairplot(df, kind='reg', diag_kind='hist', height=height, aspect=aspect,
                 plot_kws={'scatter_kws': {'color': 'purple'}, 'line_kws': {'color': 'black'}})

def annotate_regression_coefficients(x, y, **kwargs):
    slope, intercept, r_value, p_value, std_err = linregress(x, y)
    ax = plt.gca()
    equation = f'y = {slope:.2f}x + {intercept:.2f}'
    label = f'{equation}\nr = {r_value:.2f}\np = {p_value:.2e}'
    ax.annotate(label, xy=(0.6, 0.9), xycoords='axes fraction', ha='left', va='center', 
                bbox=dict(boxstyle='round,pad=0.5', fc='white', alpha=0.5))

g.map_lower(annotate_regression_coefficients)
plt.show()
```
